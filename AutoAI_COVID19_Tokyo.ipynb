```python
# CSVデータのDataFrameへのロード
import pandas as pd
df_covid19_raw = pd.read_csv('130001_tokyo_covid19_patients.csv')
display(df_covid19_raw.head())
display(df_covid19_raw.tail())
```

```python
#時系列用データ作成
df_covid19_timeseriesdata = pd.DataFrame({'Date':list(df_covid19_raw.groupby('公表_年月日').size().index), 'Number_Of_People':df_covid19_raw.groupby('公表_年月日').size().tolist()})
df_covid19_timeseriesdata
```
